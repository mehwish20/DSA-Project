# Data Visualization

Gender

```python
f, ax = plt.subplots(figsize=(4, 4))
plt.pie(data['sex'].value_counts().tolist(), 
        labels=['Female', 'Male'], colors=['#ffd1df', '#a2cffe'], 
        autopct='%1.1f%%', startangle=90)
axis = plt.axis('equal')
```

Age

```python
fig, ax = plt.subplots(figsize=(5, 4))
sns.distplot(data['age'],  
             hist_kws={"alpha": 1, "color": "#a2cffe"}, 
             kde=False, bins=8)
ax = ax.set(ylabel="Count", xlabel="Age")
```

```python
b = sns.countplot('age',hue='sex', data=data)
b.axes.set_title('Number of students in different age groups',fontsize=30)
b.set_xlabel("Age",fontsize=30)
b.set_ylabel("Count",fontsize=20)
plt.show()
```

Weekly Study Time

```python
f, ax = plt.subplots(figsize=(4, 4))
plt.pie(data['studytime'].value_counts().tolist(), 
        labels=['2 to 5 hours', '<2 hours', '5 to 10 hours', '>10 hours'], 
        autopct='%1.1f%%', startangle=0)
axis = plt.axis('equal')
```

Absences

```python
sns.histplot(data['absences'], bins=20)
plt.title('Absences distribtuion')
plt.show()
```

Initial Grade(G1) Distribution

```python
sns.histplot(data['G1'], discrete=True)
plt.title('G1 distribtuion')
plt.show()
```

Middle Grade(G2) Distribution

```python
sns.histplot(data['G2'], discrete=True)
plt.title('G2 distribtuion')
plt.show()
```

Grades according to the number of students who scored them

```python
plt.subplots(figsize=(4,8))
grade_counts = data['G3'].value_counts().sort_values().plot.barh(width=.9,color=sns.color_palette('inferno',40))
grade_counts.axes.set_title('Number of students who scored a particular grade',fontsize=30)
grade_counts.set_xlabel('Number of students', fontsize=30)
grade_counts.set_ylabel('Final Grade', fontsize=30)
plt.show()
```

Final Grade Distribution

```python
b = sns.countplot(data['G3'])
b.axes.set_title('Distribution of Final grade of students', fontsize = 30)
b.set_xlabel('Final Grade', fontsize = 20)
b.set_ylabel('Count', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,6))
for age, grouped_data in data.groupby('age'):
    if age <= 19:
        sns.kdeplot(grouped_data['G3'], label=age)
plt.legend()
plt.title('Grade distribtuion depending on age of students')
plt.show()
```

```python
sns.kdeplot(data=data, x="G3", hue="sex")
plt.title('Influence of gender on Final Grade(G3)')
plt.show()
```
